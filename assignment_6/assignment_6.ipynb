```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('data.csv')
```

```python
# 1. Pie Chart

dataset['RainToday'].value_counts().plot(kind='pie')
```

```python
# 2. Scatter Plot

sns.scatterplot(x="MinTemp", y="MaxTemp", data=dataset)
```

```python
# 3. Bubble Plot

dataset.plot.scatter(x='MinTemp', y='MaxTemp', s=dataset['Rainfall'] * 2);
```

```python
# 4. Histogram

dataset.plot(y='Rainfall', kind='hist')
```

```python
# 5. Bar Chart

dataset.plot.bar(y='Rainfall', rot=0, figsize=(100, 100))
```

```python
# 6. Line Chart

sns.lineplot(x="MinTemp", y="MaxTemp", data=dataset)
```

```python
# 7. Normal Curve

sns.distplot(dataset['MinTemp'], rug=True, hist=False)
```

```python
# 8. Horizontal Bar Graph

dataset.plot.barh(x='MinTemp', y='MaxTemp', figsize=(100, 100))
```

```python
# 9. Joint Plot

sns.jointplot(x='MinTemp', y='Rainfall', data=dataset)
```

```python
# 10. Hexbin Plot

sns.jointplot(x = dataset['MinTemp'], y = dataset['MaxTemp'], kind = 'hex')
```

```python
# 11. Kernel Density Estimation

sns.jointplot(x='Evaporation', y='Sunshine', data=dataset, kind='kde')
```

```python
# 12. Strip Plot

sns.stripplot(x = 'Humidity3pm' ,y = 'Pressure9am', data=dataset)
```

```python
# 13. Swarm Plot

sns.swarmplot(x='Rainfall', y='Sunshine', data=dataset)
```

```python
# 14. Box Plot

sns.boxplot(x='Rainfall', y='Sunshine', data=dataset)
```

```python
# 15. Column Chart

sns.barplot(x='Rainfall', y='Sunshine', data=dataset, hue='RainToday')
```

```python
# 16. Count Plot

sns.countplot(x='RainToday', data=dataset, palette='Greens_d')
```

```python
# 17. Stacked Area Chart

plt.stackplot(dataset['Sunshine'], dataset['Rainfall'])
plt.show()
```

```python
# 18. Violin Plot

sns.violinplot( y=dataset['Sunshine'], x=dataset['RainToday'], width=1)
```

```python
# 19. Multiple Line Plot

sns.lineplot(data=dataset, x='Sunshine', y='MinTemp', hue='Rainfall')
```

```python
# 20. Density Plot for multiple variables

p1=sns.kdeplot(dataset['MinTemp'], shade=True, color="r")
p1=sns.kdeplot(dataset['MaxTemp'], shade=True, color="b")
plt.show()
```
